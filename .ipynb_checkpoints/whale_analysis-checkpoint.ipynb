```python
# Initial imports
import pandas as pd
import numpy as np
import datetime as dt
from pathlib import Path

%matplotlib inline
```

# Data Cleaning

In this section, you will need to read the CSV files into DataFrames and perform any necessary data cleaning steps. After cleaning, combine all DataFrames into a single DataFrame.

Files:

* `whale_returns.csv`: Contains returns of some famous "whale" investors' portfolios.

* `algo_returns.csv`: Contains returns from the in-house trading algorithms from Harold's company.

* `sp_tsx_history.csv`: Contains historical closing prices of the S&P TSX 60 Index.

## Whale Returns

Read the Whale Portfolio daily returns and clean the data.

```python
whale_csv = Path("Resources/whale_returns.csv")

whale_df = pd.read_csv(
    whale_csv, index_col="Date", infer_datetime_format=True, parse_dates=True
)
```

```python
# Reading whale returns
whale_df.head()
```

```python
# Count nulls
whale_df.isnull().sum()
```

```python
# Drop nulls
whale_df = whale_df.dropna()
whale_df.isnull().sum()
```

## Algorithmic Daily Returns

Read the algorithmic daily returns and clean the data.

```python
# Reading algorithmic returns
algo_csv = Path("Resources/algo_returns.csv")

algo_df = pd.read_csv(
    algo_csv, index_col="Date", infer_datetime_format=True, parse_dates=True
)
```

```python
# Count nulls
algo_df.isnull().sum()
```

```python
# Drop nulls
algo_df = algo_df.dropna()
algo_df.isnull().sum()
```

## S&P TSX 60 Returns

Read the S&P TSX 60 historic closing prices and create a new daily returns DataFrame from the data. 

```python
# Reading S&P TSX 60 Closing Prices
sp_tsx_h_csv = Path("Resources/sp_tsx_history.csv")


sp_tsx_df = pd.read_csv(
    sp_tsx_h_csv, index_col="Date", infer_datetime_format=True, parse_dates=True
)
```

```python
# Check Data Types
sp_tsx_df.dtypes
```

```python
# Fix Data Types
sp_tsx_df["Close"] = sp_tsx_df["Close"].str.replace("$", "")
sp_tsx_df["Close"] = sp_tsx_df["Close"].str.replace(",", "")
sp_tsx_df["Close"] = sp_tsx_df["Close"].astype("float")
```

```python
# Calculate Daily Returns
sp_tsx_rtns = sp_tsx_df.pct_change().copy()
```

```python
# Drop nulls
sp_tsx_rtns = sp_tsx_rtns.dropna()
sp_tsx_rtns.isnull().sum()
```

```python
# Rename `Close` Column to be specific to this portfolio.
sp_tsx_rtns.rename(columns={"Close" : "S&P_TSX"}, inplace = True)
```

```python
sp_tsx_rtns.head()
```

## Combine Whale, Algorithmic, and S&P TSX 60 Returns

```python
# Join Whale Returns, Algorithmic Returns, and the S&P TSX 60 Returns into a single DataFrame with columns for each portfolio's returns.
portfolio_df = pd.concat([whale_df, algo_df, sp_tsx_rtns], axis="columns", join="inner")
portfolio_df.head()
```

# Conduct Quantitative Analysis

In this section, you will calculate and visualize performance and risk metrics for the portfolios.

## Performance Anlysis

#### Calculate and Plot the daily returns.

```python
# Plot daily returns of all portfolios
portfolio_df.plot(figsize = (20,10), title = 'Portfolio Daily Returns')
```

#### Calculate and Plot cumulative returns.

```python
# Calculate cumulative returns of all portfolios
cumulative_returns = (1 + portfolio_df).cumprod()
# Plot cumulative returns
cumulative_returns.plot(figsize = (20,10), title = 'Portfolio Cumulative Returns')
```

## Risk Analysis

Determine the _risk_ of each portfolio:

1. Create a box plot for each portfolio. 
2. Calculate the standard deviation for all portfolios.
4. Determine which portfolios are riskier than the S&P TSX 60.
5. Calculate the Annualized Standard Deviation.

### Create a box plot for each portfolio


```python
# Box plot to visually show risk
cumulative_returns.plot.box(figsize = (20,10), title = 'Portfolio Returns Box Plot')
```

### Calculate Standard Deviations

```python
# Calculate the daily standard deviations of all portfolios
portfolio_std = portfolio_df.std()
portfolio_std = portfolio_std.sort_values(ascending=False)
portfolio_std
```

### Determine which portfolios are riskier than the S&P TSX 60

```python
# Calculate  the daily standard deviation of S&P TSX 60
sp_tsx_std = portfolio_std['S&P_TSX']
print(f'The daily standard deviation of S&P TSX 60 is {sp_tsx_std}.')
# Determine which portfolios are riskier than the S&P TSX 60
risk_by_std = portfolio_std[portfolio_std>sp_tsx_std]
print(f'The following portfolios are riskier than S&P TSX 60 based on standard deviation')
print(risk_by_std)
```

### Calculate the Annualized Standard Deviation

```python
# Calculate the annualized standard deviation (252 trading days)
portfolio_ann_std = portfolio_std * np.sqrt(252)
portfolio_ann_std
```

## Rolling Statistics

Risk changes over time. Analyze the rolling statistics for Risk and Beta. 

1. Calculate and plot the rolling standard deviation for the S&P TSX 60 using a 21-day window.
2. Calculate the correlation between each stock to determine which portfolios may mimick the S&P TSX 60.
3. Choose one portfolio, then calculate and plot the 60-day rolling beta for it and the S&P TSX 60.

### Calculate and plot rolling `std` for all portfolios with 21-day window

```python
# Calculate the rolling standard deviation for all portfolios using a 21-day window
std_roll_21 = portfolio_df.rolling(window=21).std()
# Plot the rolling standard deviation
std_roll_21.plot(figsize = (20,10), title = 'Portfolio Rolling 21 day STD')
```

### Calculate and plot the correlation

```python
# Calculate the correlation
portfolio_corr = portfolio_df.corr()
# Display de correlation matrix
portfolio_corr
```

```python
#visualisation of correlation
import seaborn as sns
sns.heatmap(portfolio_corr, vmin = -1, vmax = 1)
```

### Calculate and Plot Beta for a chosen portfolio and the S&P 60 TSX

```python
# Calculate covariance of a single portfolio
algo2_covar_roll60 = portfolio_df['Algo 2'].rolling(window=60).cov(portfolio_df['S&P_TSX'])
# Calculate variance of S&P TSX
sp_var_roll60 = portfolio_df['S&P_TSX'].rolling(window=60).var()
# Computing beta
algo2_beta_roll60 = algo2_covar_roll60 / sp_var_roll60
# Plot beta trend
algo2_beta_roll60.plot(figsize = (20,10), title = 'Algo2 vs S&P TSX Rolling 60 day Beta')
```

## Rolling Statistics Challenge: Exponentially Weighted Average 

An alternative way to calculate a rolling window is to take the exponentially weighted moving average. This is like a moving window average, but it assigns greater importance to more recent observations. Try calculating the [`ewm`](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.ewm.html) with a 21-day half-life.

```python
# Use `ewm` to calculate the rolling window
ewm_pd = pd.DataFrame(portfolio_df).ewm(span=21).mean()
ewm_pd.plot(figsize = (20,10), title = 'Portfolio 21 day Exponentially Weighted Average')
```

# Sharpe Ratios
In reality, investment managers and thier institutional investors look at the ratio of return-to-risk, and not just returns alone. After all, if you could invest in one of two portfolios, and each offered the same 10% return, yet one offered lower risk, you'd take that one, right?

### Using the daily returns, calculate and visualize the Sharpe ratios using a bar plot

```python
# Annualized Sharpe Ratios
sharpe_ratios = (portfolio_df.mean() * 252) / (portfolio_df.std() * np.sqrt(252))
sharpe_ratios
```

```python
# Visualize the sharpe ratios as a bar plot
sharpe_ratios.plot(figsize = (10,5), kind="bar", title="Portfolio Sharpe Ratios")
```

### Determine whether the algorithmic strategies outperform both the market (S&P TSX 60) and the whales portfolios.

Based on the sharpe ratios, both Algo1 and Algo2 out performed all portfolios (including S&P TSX 60). There is one exception to this where Berkshire outperformed Algo2 slightly.

## Choose 3-5 custom stocks with at last 1 year's worth of historic prices and create a DataFrame of the closing prices and dates for each stock.

For this demo solution, we fetch data from three companies listes in the S&P TSX 60 index.

* `SHOP` - [Shopify Inc](https://en.wikipedia.org/wiki/Shopify)

* `OTEX` - [Open Text Corporation](https://en.wikipedia.org/wiki/OpenText)

* `L` - [Loblaw Companies Limited](https://en.wikipedia.org/wiki/Loblaw_Companies)

```python
# Reading data from 1st stock
flt_csv = Path("Resources/Custom_Portfolio/FLT.csv")

flt_df = pd.read_csv(
    flt_csv, index_col="Date", infer_datetime_format=True, parse_dates=True
)
flt_df = flt_df.rename(columns = {'Close' : 'FLT'})
flt_df.tail()
```

```python
# Reading data from 2nd stock
ckf_csv = Path("Resources/Custom_Portfolio/ckf.csv")

ckf_df = pd.read_csv(
    ckf_csv, index_col="Date", infer_datetime_format=True, parse_dates=True
)
ckf_df = ckf_df.rename(columns = {'Close' : 'CKF'})
ckf_df.tail()
```

```python
# Reading data from 3rd stock
wow_csv = Path("Resources/Custom_Portfolio/wow.csv")

wow_df = pd.read_csv(
    wow_csv, index_col="Date", infer_datetime_format=True, parse_dates=True
)
wow_df = wow_df.rename(columns = {'Close' : 'WOW'})
wow_df.tail()
```

```python
# Combine all stocks in a single DataFrame
custom_df = pd.concat([flt_df, ckf_df, wow_df], axis="columns", join="inner")
custom_df.head()
```

```python
# Reset Date index
print('Date index set when importing csv')
```

```python
# Reorganize portfolio data by having a column per symbol
print('Column names updated on previous steps')
```

```python
# Calculate daily returns
custom_daily = custom_df.pct_change()
# Drop NAs
custom_daily = custom_daily.dropna()
#custom_returns.isnull().sum()
# Display sample data
custom_daily.head()
```

## Calculate the weighted returns for the portfolio assuming an equal number of shares for each stock

```python
# Set weights
weights = [1/3, 1/3, 1/3]

# Calculate portfolio return
custom_returns = custom_daily.dot(weights)
# Display sample data
custom_returns = custom_returns.to_frame()
custom_returns = custom_returns.rename(columns = {0 : "Custom_Portfolio"})
custom_returns.head()
```

## Join your portfolio returns to the DataFrame that contains all of the portfolio returns

```python
# Join your returns DataFrame to the original returns DataFrame
all_df = pd.concat([portfolio_df,custom_returns], axis="columns", join="inner")
all_df.head()
```

```python
# Only compare dates where return data exists for all the stocks (drop NaNs)
all_df = all_df.dropna()
all_df.isnull().sum()
```

## Re-run the risk analysis with your portfolio to see how it compares to the others

### Calculate the Annualized Standard Deviation

```python
# Calculate the annualized `std`
custom_annual_std = all_df.std() * np.sqrt(252)
custom_annual_std
```

### Calculate and plot rolling `std` with 21-day window

```python
# Calculate rolling standard deviation
df_plot = all_df['Custom_Portfolio'].rolling(window=21).std()
# Plot rolling standard deviation
df_plot.plot(figsize = (10,5), title="Portfolio Rolling 21 day STD")
```

### Calculate and plot the correlation

```python
# Calculate and plot the correlation
correlation = all_df.corr()
sns.heatmap(correlation)
```

### Calculate and Plot the 60-day Rolling Beta for Your Portfolio compared to the S&P 60 TSX

```python
# Calculate and plot Beta
rolling_custom_covariance = all_df['Custom_Portfolio'].rolling(window=60).cov(all_df['S&P_TSX'])
rolling_variance = all_df['S&P_TSX'].rolling(window=60).var()

rolling_custom_beta = rolling_custom_covariance / rolling_variance

rolling_custom_beta.plot(figsize=(10, 5), title=
                               "Rolling 60-Day Beta Custom vs S&P TSX")
```

### Using the daily returns, calculate and visualize the Sharpe ratios using a bar plot

```python
# Calculate Annualzied Sharpe Ratios
sharpe_ratios = (all_df.mean() * 252) / (custom_annual_std)
sharpe_ratios
```

```python
# Visualize the sharpe ratios as a bar plot
sharpe_ratios.plot(figsize=(10, 5), kind='bar', title=
                               "Annualised Sharpe Ratios by portfolios")
```
